```python
import matplotlib.pyplot as plt
import numpy as np
import random

def classic_kelly(p):
    ending_vals = []
    plt.figure(p)
    p = p/100
    for _ in range(100):
        x = 100
        x_list = []
        for i in range(50):
            bet_size = 2*p - 1
            r = random.random() #r is a random value between 0 and 1
            if r < p:
                x *= 1 + bet_size
            else:
                x *= 1 - bet_size
            x_list.append(x)
        plt.plot(np.arange(len(x_list)), x_list)
        plt.xlabel('ith bet for p = ' + str(p))
        plt.ylabel('x value')
        ending_vals.append(x)
    median = np.median(ending_vals)
    mean = np.mean(ending_vals)
    print()
    print("X starts at 100. Median ending value when win probability is", p, ":  ", median)
    print("mean: ", mean)
    count = 0
    for v in ending_vals:
        if v < 50:
            count += 1
    print("'risk of ruin', or probability that we end up with < 50: ", count/100)
    return median, mean, count/100

def minus50(p):
    ending_vals = []
    plt.figure(p)
    p = p/100
    for _ in range(100):
        x = 100
        x_list = []
        for i in range(50):
            bet_size = p-0.5
            r = random.random() #r is a random value between 0 and 1
            if r < p:
                x *= 1 + bet_size
            else:
                x *= 1 - bet_size
            x_list.append(x)
        plt.plot(np.arange(len(x_list)), x_list)
        plt.xlabel('ith bet for p = ' + str(p))
        plt.ylabel('x value')
        ending_vals.append(x)
    median = np.median(ending_vals)
    mean = np.mean(ending_vals)
    print()
    print("X starts at 100. Median ending value when win probability is", p, ":  ", median)
    print("mean: ", mean)
    count = 0
    for v in ending_vals:
        if v < 50:
            count += 1
    print("'risk of ruin', or probability that we end up with < 50: ", count/100)
    return median, mean, count/100

def fixed50(p):
    ending_vals = []
    plt.figure(p)
    for _ in range(100):
        x = 100
        x_list = []
        for i in range(50):
            bet_size = 50/100
            r = random.random() #r is a random value between 0 and 1
            if r < p:
                x *= 1 + bet_size
            else:
                x *= 1 - bet_size
            x_list.append(x)
        plt.plot(np.arange(len(x_list)), x_list)
        plt.xlabel('ith bet for p = ' + str(p))
        plt.ylabel('x value')
        ending_vals.append(x)
    median = np.median(ending_vals)
    mean = np.mean(ending_vals)
    print()
    print("X starts at 100. Median ending value when win probability is", p, ":  ", median)
    print("mean: ", mean)
    count = 0
    for v in ending_vals:
        if v < 50:
            count += 1
    print("'risk of ruin', or probability that we end up with < 50: ", count/100)
    return median, mean, count/100

for p in [51, 55, 60, 75]:
    classic_kelly(p)
    #minus50(p)
    #fixed(p)
```

```python
def classic_kelly_noprint(p):
    ending_vals = []
    p = p/100
    for _ in range(100):
        x = 100
        x_list = []
        for i in range(50):
            bet_size = 2*p - 1
            r = random.random() #r is a random value between 0 and 1
            if r < p:
                x *= 1 + bet_size
            else:
                x *= 1 - bet_size
            x_list.append(x)
        ending_vals.append(x)
        
    median = np.median(ending_vals)
    mean = np.mean(ending_vals)
    count = 0
    for v in ending_vals:
        if v < 50:
            count += 1
    return median, mean, count/100

def fixed(amount):
    ending_vals = []
    for _ in range(100):
        x = 100
        x_list = []
        for i in range(50):
            r = random.random() #r is a random value between 0 and 1
            if r < p:
                x += amount
            else:
                x -= amount
            x_list.append(x)
        ending_vals.append(x)
        
    median = np.median(ending_vals)
    mean = np.mean(ending_vals)
    count = 0
    for v in ending_vals:
        if v < 50:
            count += 1
    return median, mean, count/100

def ratio(percentage):
    ending_vals = []
    for _ in range(100):
        x = 100
        x_list = []
        for i in range(50):
            bet_size = percentage/100
            r = random.random() #r is a random value between 0 and 1
            if r < p:
                x *= 1 + bet_size
            else:
                x *= 1 - bet_size
            x_list.append(x)
        ending_vals.append(x)
        
    median = np.median(ending_vals)
    mean = np.mean(ending_vals)
    count = 0
    for v in ending_vals:
        if v < 50:
            count += 1
    return median, mean, count/100

def minus50func(p):
    ending_vals = []
    for _ in range(100):
        x = 100
        x_list = []
        for i in range(50):
            bet_size = p-50
            r = random.random() #r is a random value between 0 and 1
            if r < p:
                x *= 1 + bet_size
            else:
                x *= 1 - bet_size
            x_list.append(x)
        ending_vals.append(x)
        
    median = np.median(ending_vals)
    mean = np.mean(ending_vals)
    count = 0
    for v in ending_vals:
        if v < 50:
            count += 1
    return median, mean, count/100

kellys = []
fixed10 = []
fixed33 = []
fixed67 = []
ratio10 = []
ratio33 = []
ratio67 = []
minus_50 = []

p_vals = [51, 55, 60, 65, 70, 75, 80, 85, 90, 95, 99]
for p in p_vals:
    kellys.append(classic_kelly_noprint(p)[0])
    fixed10.append(fixed(10)[0])
    fixed33.append(fixed(33)[0])
    fixed67.append(fixed(67)[0])
    ratio10.append(ratio(10)[0])
    ratio33.append(ratio(33)[0])
    ratio67.append(ratio(67)[0])
    minus_50.append(minus50func(p)[0])
    
plt.plot(np.arange(len(p_vals)), kellys)
plt.plot(np.arange(len(p_vals)), fixed10)
plt.plot(np.arange(len(p_vals)), fixed33)
plt.plot(np.arange(len(p_vals)), fixed67)
plt.plot(np.arange(len(p_vals)), ratio10)
plt.plot(np.arange(len(p_vals)), ratio33)
plt.plot(np.arange(len(p_vals)), ratio67)
plt.plot(np.arange(len(p_vals)), minus_50)
plt.legend(["kelly", "fixed10", "fixed30", "fixed67", "ratio10", "ratio33", "ratio67", "minus-50"])
```

```python
def uniform(percentage):
    ending_vals = []
    plt.figure(percentage)
    for _ in range(1000):
        x = 100
        x_list = []
        for i in range(20):
            bet = x*percentage/100
            r = 3*random.random() - 1 #r is a random value between -1 and 2
            x = x + r*bet #these two lines can be simplified but this is clearer
            x_list.append(x)
        plt.plot(np.arange(len(x_list)), x_list)
        plt.xlabel('ith bet of size ' + str(percentage) + '%')
        plt.ylabel('x value')
        ending_vals.append(x_list[len(x_list) - 1])
    print()
    print("X starts at 100. Median ending value when ", percentage, "%  is bet each time: ", np.median(ending_vals))
    count = 0
    for v in ending_vals:
        if v < 50:
            count += 1
    print("'risk of ruin', or probability that we end up with < 50: ", count/1000)

for p in [10, 25, 50, 75, 99]:
    uniform(p)
```
